```python
# 판다스 import
import pandas as pd
```

```python
# 판다스 버전 확인
print(pd.__version__)
```

```python
# 문자열을 원소로 갖는 1차원 구조의 리스트 객체 생성
data1 = ['a', 'b', 'c', 'd', 'e']
```

```python
# 리스트 객체 출력 및 타입 확인
print(data1)
print("자료형: ", type(data1))
```

```python
# 리스트 객체를 판다스 시리즈로 변환
sr1 = pd.Series(data1)
print("자료형: ", type(sr1))
```

```python
# 판다스 시리즈 출력
print(sr1)
```

```python
# 시리즈 객체 원소 추출
sr1.loc[0]
```

```python
# 인덱스 범위를 지정하여 슬라이싱 추출
sr1.loc[1:3]
```

```python
# 숫자 데이터로 구성된 튜플 형태의 1차원 벡터를 시리즈로 변환
data2 = (1, 2, 3.14, 100, -10)
sr2 = pd.Series(data2)
print(sr2)  # 정수형 데이터의 dtype은 int64
```

```python
# 시리즈를 결합하여 데이터프레임 만들기
dict_data = {'c0':sr1, 'c1':sr2}  # 이렇게 딕셔너리를 생성할 경우 key값을 컬럼명, value값을 데이터로 하는 데이터 프레임이 생성됨
df1 = pd.DataFrame(dict_data)
df1
```

```python
# 데이터프레임의 자료형 확인
type(df1)
```

```python
# 데이터프레임의 열 이름 추출
df1.columns
```

```python
# 데이터프레임 컬럼명 변경
df1.columns = ['string', 'number']
df1
```

```python
# 데이터프레임에서 행 인덱스 추출
df1.index
```

```python
# 행 인덱스 수정
df1.index = ['r0', 'r1', 'r2', 'r3', 'r4']
df1
```

```python
# 데이터프레임 일부분 추출 loc[행 인덱스, 열 이름] - 단일 데이터 추출
df1.loc['r2', 'number']
```

```python
# 데이터프레임 일부분 범위로 추출 - 데이터프레임 추출
df1.loc['r2':'r3', 'string':'number']
```

```python
# 데이터프레임 일부분 범위로 컬럼 일부 추출 - 시리즈 추출
tempSeries1 = df1.loc['r2':'r3', 'number']
tempSeries1
```

```python
# 데이터프레임 컬럼 일부를 추출한 시리즈 타입 확인
type(tempSeries1)
```

```python
# 데이터프레임 일부분 범위로 로우 일부 추출 - 시리즈 추출
tempSeries2 = df1.loc['r2', 'string':'number']
tempSeries2
```

```python
# 데이터프레임 로우 일부를 추출한 시리즈 타입 확인
type(tempSeries2)
```

```python
# 데이터프레임 일부분 범위로 컬럼 전체 추출 - 시리즈 추출
df1.loc[:, 'string']
```

```python
# 데이터프레임 일부분 범위로 데이터프레임 일부 추출
df1.loc['r2':'r3', :]
```
